```python
#a. Print the list of 20 random numbers generated between 1 and 100
import random
randomnumberlist=[]
for i in range(20):
    i=random.randint(1,100)
    randomnumberlist.append(i)
print(randomnumberlist)
```

```python
#b. Average of the elements in the list
averageoflist=sum(randomnumberlist)/len(randomnumberlist)
print(averageoflist)
```

```python
#c. Largest and smallest values in the list
print(max(randomnumberlist))
print(min(randomnumberlist))
```

```python
#d. Second largest and the second smallest element in the list
randomnumberlist.sort()
print(randomnumberlist[1])
print(randomnumberlist[-2])
```

```python
#e. Print how many even numbers in the list
evenlist=[i for i in randomnumberlist if i%2==0]
len(evenlist)
```

```python
list12=[]
newlist=[]
for i in range(10):
    list12.append(random.randint(0,1))
print(list12)
for i in range(1,len(list12)):
    for j in range(i):
        if j<i and list12[i]==list12[j]:
            newlist.extend([list12[i],list12[j]])
        else:
            newlist=[]
print(newlist)
```
