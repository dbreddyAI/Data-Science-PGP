```python
number = eval(input("Enter a number: "))
sumof=0
for i in range(1,number+1):
    if number%i ==0:
        sumof +=i
print(f"The sum of the divisors of the number {number} is {sumof}")
```

```python
sumof=0
for number in range(1,10001):
    for j in range(1,number):
        if number%j==0:
            sumof +=j
    if sumof == number:
        print(number)
    sumof=0
```

```python
import math
li =[]
number = eval(input("Enter a number: "))
for i in range(1,number):
    if number%i==0 and i%math.sqrt(i)==0:
            li.append(i)
if len(li)>1:
    print("the number is not square free")
else:
    print("The number is squrare free")
```

```python
sq,cu,fi = 0,0,0
for i in range(1,1001):
    if i%i**(0.5)==0:
        sq+=1
    elif i%i**(1/3)==0:
        cu+=1
    elif i%i**(1/5)==0:
        fi+=1
    else:
        continue
print(f"Below 1000 there are {sq} perfect squares,{cu} perfect cubes,{fi} perfect fifth powers")
```

```python
# A Simple Method to count cubes between a and b 

def printCubes(a, b) : 
	# Traverse through all numbers in given range 
	# and one by one check if number is prime 
	for i in range(a, b + 1) : 
		
		# Check if current number 'i' 
		# is perfect cube 
		j = 1
		for j in range(j, i + 1 ) : 
			
			if (j ** 3 == i) : 
				print( j ** 3, end = " ") 
				break
			

# Driver code 

a = 1; b = 1000
print("Perfect cubes in given range: ") 
printCubes(a, b) 


# This code is contributed by Nikita Tiwari. 
```
