```python
count=0
for i in range(1,101):
    if (i**2)%10 ==1:
#         print(i**2)
        count+=1
print(f'The number of squares that end with 1 is {count}')
```

```python
count4=0
count9=0
for i in range(1,101):
    if (i**2)%10 ==4:
        print(i**2)
        count4+=1
    if (i**2)%10 ==9:
        print(i**2)
        count9+=1
print(f'The number of squares that end with 4 is {count4}')
print(f'The number of squares that end with 9 is {count9}')
```
