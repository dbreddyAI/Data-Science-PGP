```python
name=input("Enter your name: ")
repeat=eval(input("How many times do you want to print your name: "))
for i in range(repeat+1):
    print(name,end=',')
```

```python
number=eval(input('Enter the number of terms you want: '))
previous,initial,next=1,1,None
print(previous,initial,sep=',',end=',')
for i in range(number-2):
    next=initial+previous
    print(next,end=',')
    previous=initial
    initial=next
```

```python
width=eval(input("Enter the width of the box: "))
height=eval(input("Enter the height of the box: "))
for i in range(height):
    print('*'*width)
```

```python
width=eval(input("Enter the width of the box: "))
height=eval(input("Enter the height of the box: "))
for i in range(height):
    if i==0 or i==height-1:
        print('*'*width)
    else:
        print('*'," "*(width-4),'*')
```

```python
number=eval(input("Enter the height of the triangle: "))
for i in range(1,number+1):
    print('*'*i)
```

```python
number=eval(input("Enter the height of the triangle: "))
for i in range(number,0,-1):
    print('*'*i)
```

```python
height=eval(input("Enter the height of the triangle: "))
width=height
for i in range(height):
    for j in range(width):
        if i==0 and j==(width/2):
            print(1,end=' ')
        else:
            print(1,end=' ')
    print()
```
