```python
time = eval(input("Enter hour: "))
section = eval(input("am (1) or pm (2)? "))
forward = eval(input("How many hours ahead?"))
if section ==1:
    if (time + forward)%12 >0:
        time = (time+forward)%12
        print(f"New Hour: {time} pm")
    else:
        time = time+forward
        print(f"New Hour: {time} am")
elif section==2:
    if (time + forward)%12 >0:
        time = (time+forward)%12
        print(f"New Hour: {time} am")
    else:
        time = time+forward
        print(f"New Hour: {time} pm")
```
